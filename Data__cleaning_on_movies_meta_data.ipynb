```python
#new_idea:- create a column for each genre-say 1if the movie has the genre else put 0
#we are going to find all the unique genres
uniqueGenres = set()
for i in data['genres']:
  genlist = ast.literal_eval(i)
  for j in genlist:
    uniqueGenres.add(j['name'])
```

```python
print(uniqueGenres)
len(uniqueGenres)
```

```python
rows = []  #for all the movies
genres = list(uniqueGenres)
for i in data['genres']: #looping through all rows in the movies column
    genlist = ast.literal_eval(i) #convert string to list of dictionary
    movieGenres = [x['name'] for x in genlist]
    row  = [] #is stores a list 1's and 0's for one movie
    for j in genres:
      row.append(int(j in movieGenres))
    rows.append(row)
```

```python
mg = ['action','adventure']
unique = ['action','adventure','family','horror']
row = []
for j in unique:
  row.append(int(j in mg))
row
```

```python
#convert the encodings in the rows in to dataframe
genre = pd.DataFrame(rows,columns=genres)
genre
```

```python
#concat the genre with data
data = pd.concat([data,genre],axis=1)
data.head()
```

```python
#drop the genres columns
data = data.drop(columns=['genres'])
data.head()
```
